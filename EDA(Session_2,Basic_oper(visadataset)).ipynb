**task-1**
- I want to create two lists
- One is categorial column list
- Second one is numerical column list
- use dtypes

- whenever you want to create any list from series 
- first convert the series into a dictionary
- when u convert it becomes a key value pair
- iterate thru dictionary
- if that value==object type,then keep that corresponding key in one list
- else keep that varaible in other list

```python
data_types=visa_df.dtypes
dict(data_types)
```

```python
cat_list=[]
num_list=[]
for i,j in dict(data_types).items():
    if j=='object':
        cat_list.append(i)
    else:
        num_list.append(i)
print(cat_list)
print(num_list)
```

```python
cat_list=[i for i,j in dict(data_types).items() if j=='object']
num_list=[i for i,j in dict(data_types).items() if j!='object']
cat_list,num_list
```

```python
cat_list=[]
num_list=[]
for i,j in dict(data_types).items():
    if j=='object':
        cat_list.append(i)
    else:
        num_list.append(i)
cat_list,num_list
```

```python
len(cat_list),len(num_list)
```

$take$
- first argument:indices
- indices will take some values in array: in a list
- it tries to extract the data
- the values we provided represents columns or rowa?
- axis=0 rep about rows
- axis=1 rep abput columns
- by default axis mentioned as zero

```python
#CASE-1
visa_df.take([100,200,300])
# it retrives 3 rows which has index 100,200,300
```

```python
#CASE-2
visa_df.take([100,200,300],axis=1)
# now this will check for 100th,200th,300th columns
# but orginally we have only 12 columns(0,11)
#we will get error
```

```python
#CASE-3:select 3 columns 2nd,5th and 7th using take
visa_df.take([2,5,7],axis=1)
#the output also behave as dataframe
```

```python
#CASE-4: slicing will fail
visa_df.take([2:8],axis=1)
```
